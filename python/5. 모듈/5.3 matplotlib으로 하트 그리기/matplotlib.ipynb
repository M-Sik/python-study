# 5.3. matplotlib으로 하트 그리기

## matplotlib으로 하트 그리기

```python
%matplotlib inline

from pylab import *

x = linspace(-1.6, 1.6, 10000)
f = lambda x: (sqrt(cos(x)) * cos(200 * x) + sqrt(abs(x)) - 0.7) * \
    pow((4 - x * x), 0.01)
plot(x, list(map(f, x)))
show()
```
